```python
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.static import teams as nba_teams_static
import pandas as pd
from datetime import datetime
import math

# Function to round a number to a specified number of significant figures
def round_sig(x, sig=4):
    if x == 0:
        return 0
    return round(x, sig - int(math.floor(math.log10(abs(x)))) - 1)

# -----------------------------
# Regular Season: Monthly Data
# -----------------------------
# Define the season parameters
season = '2023-24'
regular_season_type = 'Regular Season'

# Retrieve regular season game data for the specified season
regular_gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable=season, season_type_nullable=regular_season_type)
regular_games = regular_gamefinder.get_data_frames()[0]

# List of NBA team names (used for filtering and ordering)
nba_teams = [
    "Atlanta Hawks", "Boston Celtics", "Brooklyn Nets", "Charlotte Hornets", "Chicago Bulls", 
    "Cleveland Cavaliers", "Dallas Mavericks", "Denver Nuggets", "Detroit Pistons", "Golden State Warriors", 
    "Houston Rockets", "Indiana Pacers", "Los Angeles Clippers", "Los Angeles Lakers", "Memphis Grizzlies", 
    "Miami Heat", "Milwaukee Bucks", "Minnesota Timberwolves", "New Orleans Pelicans", "New York Knicks", 
    "Oklahoma City Thunder", "Orlando Magic", "Philadelphia 76ers", "Phoenix Suns", "Portland Trail Blazers", 
    "Sacramento Kings", "San Antonio Spurs", "Toronto Raptors", "Utah Jazz", "Washington Wizards"
]

# Define month ranges for the season.
# Each tuple: (Month Name, Start Date, Exclusive End Date)
month_ranges = [
    ('October', datetime(2023, 10, 1), datetime(2023, 11, 1)),
    ('November', datetime(2023, 11, 1), datetime(2023, 12, 1)),
    ('December', datetime(2023, 12, 1), datetime(2024, 1, 1)),
    ('January', datetime(2024, 1, 1), datetime(2024, 2, 1)),
    ('February', datetime(2024, 2, 1), datetime(2024, 3, 1)),
    ('March', datetime(2024, 3, 1), datetime(2024, 4, 1)),
    ('April', datetime(2024, 4, 1), datetime(2024, 5, 1))
]

# Dictionary to hold monthly win percentages
monthly_win_percentages = {}

for month_name, start_date, end_date in month_ranges:
    # Filter games for the current month and only for teams in our NBA list
    month_games = regular_games[
        (pd.to_datetime(regular_games['GAME_DATE']) >= start_date) &
        (pd.to_datetime(regular_games['GAME_DATE']) < end_date) &
        (regular_games['TEAM_NAME'].isin(nba_teams))
    ]
    
    # Calculate wins and total games by team for the month
    month_wins = month_games[month_games['WL'] == 'W'].groupby('TEAM_NAME').size()
    month_total_games = month_games.groupby('TEAM_NAME').size()
    
    # Compute win percentage (if a team played no games, fill with 0)
    win_pct = (month_wins / month_total_games).fillna(0) * 100
    win_pct = win_pct.reindex(nba_teams, fill_value=0)
    
    monthly_win_percentages[month_name] = win_pct

# Create a DataFrame with rows as teams and columns as months
monthly_win_percentage_df = pd.DataFrame(monthly_win_percentages)

# -----------------------------
# Postseason: Playoff Performance
# -----------------------------
# Retrieve playoff game data for the same season
post_season_type = 'Playoffs'
post_gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable=season, season_type_nullable=post_season_type)
post_games = post_gamefinder.get_data_frames()[0]

# Filter playoff games to include only our NBA teams
post_games = post_games[post_games['TEAM_NAME'].isin(nba_teams)]

# Calculate playoff wins and total games for each team
post_wins = post_games[post_games['WL'] == 'W'].groupby('TEAM_NAME').size()
post_total_games = post_games.groupby('TEAM_NAME').size()

# Compute playoff win percentage; teams with no playoff games get 0%
post_win_pct = (post_wins / post_total_games).fillna(0) * 100
post_win_pct = post_win_pct.reindex(nba_teams, fill_value=0)

# Also, compute total playoff wins for each team
total_playoff_wins = post_wins.reindex(nba_teams, fill_value=0)

# -----------------------------
# Merge and Export Data
# -----------------------------
# Add postseason metrics as new columns to the regular season DataFrame
monthly_win_percentage_df['Postseason_Win_Pct'] = post_win_pct
monthly_win_percentage_df['Total_Playoff_Wins'] = total_playoff_wins

# Sort teams by Total_Playoff_Wins (descending) then by Postseason_Win_Pct (descending)
final_df = monthly_win_percentage_df.sort_values(by=['Total_Playoff_Wins', 'Postseason_Win_Pct'], ascending=False)

# Round all numeric values to four significant figures using DataFrame.apply with Series.map
final_df = final_df.apply(lambda col: col.map(lambda x: round_sig(x, 4) if isinstance(x, (int, float)) else x))

# Write the final DataFrame to a CSV file (with team names as row index)
final_df.to_csv('nba_team_performance.csv', index_label='Team')

print("CSV file 'nba_team_performance.csv' created with teams sorted by playoff performance (total wins and win percentage).")
```

```python
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.static import teams as nba_teams_static
import pandas as pd

# Define season parameters for the 2023-2024 postseason
season = "2023-24"
post_season_type = "Playoffs"

# Retrieve playoff game data for the season
post_gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable=season, season_type_nullable=post_season_type)
post_games = post_gamefinder.get_data_frames()[0]

# Get the list of all NBA team full names
all_teams = [team['full_name'] for team in nba_teams_static.get_teams()]

# Calculate playoff wins and total playoff games for teams present in the playoff data
post_wins = post_games[post_games['WL'] == 'W'].groupby('TEAM_NAME').size()
post_total_games = post_games.groupby('TEAM_NAME').size()

# Calculate playoff win percentage; teams not in the playoffs will get 0%
playoff_win_pct = (post_wins / post_total_games).fillna(0) * 100

# Build a ranking DataFrame for all NBA teams
ranking_df = pd.DataFrame(index=all_teams)
ranking_df.index.name = 'Team'
ranking_df['Total_Playoff_Wins'] = post_wins.reindex(all_teams, fill_value=0)
ranking_df['Playoff_Win_Pct'] = playoff_win_pct.reindex(all_teams, fill_value=0)

# Sort teams: first by Total Playoff Wins (descending), then by Playoff Win Percentage (descending)
ranking_df = ranking_df.sort_values(by=['Total_Playoff_Wins', 'Playoff_Win_Pct'], ascending=False)

print("NBA Teams ranked by playoff performance (Season {}):".format(season))
print(ranking_df.to_string())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Read the CSV file (assumes "Team" is the index column and necessary columns are present)
df = pd.read_csv('nba_team_performance.csv', index_col='Team')

# Define predictor columns (months) and the target variable (postseason win percentage)
months = ["October", "November", "December", "January", "February", "March", "April"]
target = "Postseason_Win_Pct"

# Drop any rows with missing values in the predictors or target
df = df.dropna(subset=months + [target])

# Create predictor matrix X and target vector y
X = df[months].values
y = df[target].values

# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Get predictions and calculate the R-squared value
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print("Model Coefficients (per month):", model.coef_)
print("Model Intercept:", model.intercept_)
print("R-squared:", r2)

# Optionally, create a DataFrame to compare actual vs. predicted postseason performance
results = pd.DataFrame({
    'Actual': y,
    'Predicted': y_pred
}, index=df.index)
print(results)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.patches as mpatches

# Read the CSV file (assumes "Team" is the index column)
df = pd.read_csv('nba_team_performance.csv', index_col='Team')
df.reset_index(inplace=True)  # now "Team" is a column

# Define the months and the target postseason performance metric column
months = ["October", "November", "December", "January", "February", "March", "April"]
postseason_col = "Postseason_Win_Pct"  # This column must exist in the CSV

# Create a sorted list of unique team names
unique_teams = sorted(df["Team"].unique())

# Define a consistent palette for the teams.
# For example, we'll use the 'tab10' palette with as many colors as teams.
palette = sns.color_palette("tab10", len(unique_teams))

# Create custom patches for a global legend (key)
patches = [mpatches.Patch(color=palette[i], label=team) for i, team in enumerate(unique_teams)]

# Set up the plotting grid
n_months = len(months)
ncols = 3  # Number of columns in grid
nrows = int(np.ceil(n_months / ncols))

plt.figure(figsize=(ncols * 5, nrows * 4))
fig = plt.gcf()  # get the current figure

# Loop through each month and create a subplot
for i, month in enumerate(months, start=1):
    ax = plt.subplot(nrows, ncols, i)
    # Plot scatter using 'Team' as hue with a consistent palette
    sns.scatterplot(data=df, x=month, y=postseason_col, hue="Team", 
                    ax=ax, s=80, palette=palette, legend=False)
    
    # Compute regression line using numpy polyfit
    x = df[month]
    y = df[postseason_col]
    coeffs = np.polyfit(x, y, 1)
    x_fit = np.linspace(x.min(), x.max(), 100)
    y_fit = np.polyval(coeffs, x_fit)
    ax.plot(x_fit, y_fit, color='black', linestyle='--')
    
    # Calculate Pearson correlation coefficient for the subplot
    corr = df[month].corr(df[postseason_col])
    
    ax.set_title(f"{month}\nCorr: {corr:.2f}")
    ax.set_xlabel(f"{month} Win %")
    ax.set_ylabel("Postseason Win %")

plt.tight_layout()

# Add a global legend (key) for teams outside the subplots
plt.figlegend(handles=patches, loc='upper right', title="Team", bbox_to_anchor=(0.95, 0.95))
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Read the CSV file (assumes "Team" is the index column)
df = pd.read_csv('nba_team_performance.csv', index_col='Team')

# Define the months and the target postseason performance metric
months = ["October", "November", "December", "January", "February", "March", "April"]
target = "Postseason_Win_Pct"

# Dictionaries to store model performance metrics and predictions
model_results = {}
predictions = {}

# For each month, build a regression model and compute predictions and error metrics
for month in months:
    X = df[[month]].values  # predictor: win percentage for the given month
    y = df[target].values   # target: postseason win percentage
    
    # Fit linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Make predictions and calculate metrics
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    
    model_results[month] = {
        "R2": r2,
        "MSE": mse,
        "Coefficient": model.coef_[0],
        "Intercept": model.intercept_
    }
    predictions[month] = y_pred

# Print model performance for each month
print("Monthly Models for Predicting Postseason Win Percentage:\n")
print("Month\tR²\t\tMSE\t\tCoefficient\tIntercept")
for month in months:
    res = model_results[month]
    print(f"{month}\t{res['R2']:.4f}\t{res['MSE']:.4f}\t{res['Coefficient']:.4f}\t\t{res['Intercept']:.4f}")

# Create a DataFrame comparing actual vs. predicted postseason win percentage for each team
pred_df = pd.DataFrame(index=df.index)
pred_df["Actual"] = df[target]
for month in months:
    pred_df[f"Predicted_{month}"] = predictions[month]

print("\nPredicted vs. Actual Postseason Win Percentage by Team:\n")
print(pred_df.to_string())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file (assumes "Team" is the index column)
df = pd.read_csv('nba_team_performance.csv', index_col='Team')

# Define the months that are available in the CSV and the target postseason metric
months = ["October", "November", "December", "January", "February", "March", "April"]
target = "Postseason_Win_Pct"  # Actual postseason performance column

# For each month, create a table showing the projected postseason results based on that month's win percentage.
for month in months:
    # Sort teams by the given month's win percentage (highest first)
    sorted_df = df.sort_values(by=month, ascending=False).copy()
    # Assign projected rank (1 is best)
    sorted_df["Projected_Rank"] = range(1, len(sorted_df) + 1)
    
    # Print the table data to console (optional)
    print(f"Projected Postseason Results Based on {month}:")
    print(sorted_df[[month, "Projected_Rank"]])
    print("\n" + "="*50 + "\n")
    
    # Create a figure and add a table to display the projected results.
    fig, ax = plt.subplots(figsize=(8, len(sorted_df)*0.3 + 1))
    ax.axis('tight')
    ax.axis('off')
    
    # Create the table using the desired columns
    table = ax.table(
        cellText=sorted_df[[month, "Projected_Rank"]].values,
        rowLabels=sorted_df.index,
        colLabels=[f"{month} Win %", "Projected Rank"],
        loc='center'
    )
    
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    ax.set_title(f"Projected Postseason Results Based on {month}", fontweight="bold")
    plt.show()

# Now create a table for the actual postseason results.
# We will rank teams based on the target column ("Postseason_Win_Pct")
sorted_actual = df.sort_values(by=target, ascending=False).copy()
sorted_actual["Actual_Rank"] = range(1, len(sorted_actual) + 1)

print("Actual Postseason Results:")
print(sorted_actual[[target, "Actual_Rank"]])
print("\n" + "="*50 + "\n")

fig, ax = plt.subplots(figsize=(8, len(sorted_actual)*0.3 + 1))
ax.axis('tight')
ax.axis('off')

actual_table = ax.table(
    cellText=sorted_actual[[target, "Actual_Rank"]].values,
    rowLabels=sorted_actual.index,
    colLabels=[f"{target}", "Actual Rank"],
    loc='center'
)
actual_table.auto_set_font_size(False)
actual_table.set_fontsize(10)
ax.set_title("Actual Postseason Results", fontweight="bold")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file (assumes "Team" is the index column)
df = pd.read_csv('nba_team_performance.csv', index_col='Team')

# Define the months (columns with monthly win percentages) and the postseason metric column
months = ["October", "November", "December", "January", "February", "March", "April"]
target = "Postseason_Win_Pct"  # Higher is better

# Create a DataFrame to store ranking results.
# Actual postseason ranking is computed from the target column (higher win% = better rank).
ranking_df = pd.DataFrame(index=df.index)

# Compute actual postseason rank (1 is best, using 'min' so ties get the same rank)
ranking_df["Actual_Rank"] = df[target].rank(ascending=False, method='min').astype(int)

# For each month, compute the projected rank based solely on that month's win percentage.
for month in months:
    ranking_df[f"{month}_Rank"] = df[month].rank(ascending=False, method='min').astype(int)

# Order teams by actual postseason rank
ranking_df = ranking_df.sort_values("Actual_Rank")

# Create a plot comparing the actual postseason rank with each month's projected rank.
plt.figure(figsize=(12, 8))
teams_order = ranking_df.index.tolist()
x = range(1, len(teams_order) + 1)

# Plot the actual postseason ranking (solid black line)
plt.plot(x, ranking_df["Actual_Rank"], label="Actual", marker='o', linestyle='-', color='black')

# Define a color map for the months (optional customization)
month_colors = {
    "October": "blue",
    "November": "orange",
    "December": "green",
    "January": "red",
    "February": "purple",
    "March": "brown",
    "April": "magenta"
}

# Plot each month's projected ranking (dashed lines)
for month in months:
    plt.plot(x, ranking_df[f"{month}_Rank"],
             label=f"{month} Projected", marker='o', linestyle='--',
             color=month_colors.get(month, "gray"))

plt.gca().invert_yaxis()  # Invert y-axis so that rank 1 is at the top
plt.xticks(x, teams_order, rotation=90)
plt.xlabel("Team (Ordered by Actual Postseason Rank)")
plt.ylabel("Rank (1 = Best)")
plt.title("Actual vs. Projected Postseason Rankings by Month")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr

# Read the CSV file (assumes "Team" is the index column)
df = pd.read_csv('nba_team_performance.csv', index_col='Team')

# Define the months and the target postseason performance metric column
months = ["October", "November", "December", "January", "February", "March", "April"]
target = "Postseason_Win_Pct"  # This column must exist in the CSV

# Compute the actual postseason ranking based on the target column.
# Higher postseason win percentage yields a better (lower-numbered) rank.
actual_rank = df[target].rank(ascending=False, method='min')

accuracy_results = {}

# For each month, compute the projected ranking and then calculate:
# - Spearman rank correlation with the actual postseason ranking.
# - Mean absolute error (MAE) between the projected and actual rankings.
for month in months:
    # Projected ranking based on that month's win percentage
    projected_rank = df[month].rank(ascending=False, method='min')
    
    # Spearman rank correlation between actual and projected rankings
    spearman_corr, _ = spearmanr(actual_rank, projected_rank)
    
    # Mean absolute error of the rankings
    mae = np.mean(np.abs(actual_rank - projected_rank))
    
    accuracy_results[month] = {"Spearman": spearman_corr, "MAE": mae}

# Print the ranking accuracy results for each month
print("Ranking Accuracy Measures (Higher Spearman and lower MAE indicate better predictive power):")
for month in months:
    print(f"{month}: Spearman Correlation = {accuracy_results[month]['Spearman']:.4f}, MAE = {accuracy_results[month]['MAE']:.4f}")
```
